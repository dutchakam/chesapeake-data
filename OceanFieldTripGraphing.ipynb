```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_fall = pd.read_excel('/Users/AlexDutchak/Desktop/NVCC2021/Fall2021Courses/GOL111/FieldTripData/OceanFall2019data.xlsx')
df_summer = pd.read_excel('/Users/AlexDutchak/Desktop/NVCC2021/Fall2021Courses/GOL111/FieldTripData/OceanSummer2015data.xlsx')
```

```python
df_fall['Season'] = 'Fall'
df_summer['Season'] = 'Summer'

df = pd.concat([df_fall, df_summer])

df
```

```python
# function for creating lineplots for each column in the dataset by water depth

def plot_chesapeake_data(df, colors, rows=5, columns=1):
    
    fig, axes = plt.subplots(rows, columns, figsize=(8, 20))
    axes = axes.flatten()
    
    for i, j, k in zip(df.columns[1:6], axes, colors):
        
        x_val = df['Depth(m)']
        y_val = df[i]
        hue = df['Season']
        color = sns.color_palette(k, 2)
        
        sns.lineplot(x=x_val, y=y_val, hue=hue, palette=color, ax=j)
        j.grid()
        
        j.set_title(f'Chesapeake Bay {i} by Depth')
      
    plt.tight_layout()
    plt.savefig('ChesapeakeData.png', bbox_inches='tight')
    
colors = ['mako', 'flare', 'crest', 'magma', 'viridis']

plot_chesapeake_data(df, colors)
```
